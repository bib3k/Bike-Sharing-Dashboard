```python
#changing the column names for ease
new_cols_name_dict = { 
 'timestamp':'Time',
    'cnt':'Count',
    't1':'Real_temp_C',
    't2':'FeelsLike_C',
    'hum':'Humidity_percent',
    'wind_speed':' Wind_Speed_Kmph',
    'weather_code':'weather',
    'is_holiday':'On_Holiday',
    'is_weekedn':'On_Weekend',
    'season':'Season'
}
bikedf.rename(new_cols_name_dict, axis=1, inplace= True)
```

```python
#Changing humidity to normalize percentage
bikedf.Humidity_percent= bikedf.Humidity_percent/100
```

```python
#changing weather and season dtype to string
bikedf['Season']= bikedf['Season'].astype('str')
bikedf['weather']=bikedf['weather'].astype('str')
```

```python
# Mapping weather and season data values to actual weather. In other words decoding weather and season data 
Weather_dict= {
    '1.0' :'Clear',
    '2.0':'scattered clouds few clouds',
    '3.0' : 'Broken clouds',
    '4.0': 'Cloudy' ,
    '7.0' : 'light Rain shower',
    '10.0' : 'rain with thunderstorm', 
    '26.0': 'snowfall',
    '94' : 'Freezing Fog'
}
Season_dict = {
    '0.0':'Spring',
    '1.0': 'Summer',
    '2.0':'Autumn',
    '3.0':'Winter'
}

```

```python
bikedf.info()
```

```python
#changing weather and season dtype to string
bikedf['Season']= bikedf['Season'].astype('str')
bikedf.Season= bikedf.Season.map(Season_dict) # mapping 
bikedf['weather']=bikedf['weather'].astype('str')
bikedf.weather=bikedf.weather.map(Weather_dict) # mapping
```
